```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
import matplotlib.pyplot as plt
import random

w = 0.5
h = 9
f = 6
polygon = Polygon([(0, 0), (w, 0), (w, h-w), (f+w, h-w), (f+w, 0), (f+2*w, 0), (f+2*w, h), (0, h)])
x,y = polygon.exterior.xy
plt.plot(x, y, "k*--")
plt.show()
```

```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
import matplotlib.pyplot as plt
import random

w = 1.5
h = 12
f = 8
polygon = Polygon([(0, 0), (w, 0), (w, h-w), (f+w, h-w), (f+w, 0), (f+2*w, 0), (f+2*w, h), (0, h)])
x,y = polygon.exterior.xy
plt.plot(x, y, "k*--")
plt.show()
```

```python
plt.plot(x, y, "k*--")
for _ in range(10):
    point = Point(random.uniform(0.,12), random.uniform(0,12))
    if polygon.contains(point):
        plt.plot(point.x, point.y, "r*")
    else:
        plt.plot(point.x, point.y, "bo")
print(polygon.contains(point))
plt.show()
```
